```python
import pandas as pd
import os

# Excel-Datei laden
df = pd.read_excel("UX-Daten/UX-Daten.xlsx")

# Daten√ºberblick ‚Äì optional
print(df.head())

# üî¢ Altersverteilung
alter_median = df["Alter"].median()
alter_min = df["Alter"].min()
alter_max = df["Alter"].max()
alter_range = alter_max - alter_min

# üß† Technische Vorerfahrung auswerten (Ja/Nein)
tech_col = "Erfahrung mit Augmented Reality"
tech_counts = df[tech_col].value_counts(normalize=True) * 100  # in Prozent

# Online-Kaufverhalten (%)
kauf_col = "Wie oft kaufst du online ein?"
kauf_counts = df[kauf_col].value_counts(normalize=True) * 100

# Tabelle aufbauen
rows = [
    ("Alter Median", f"{alter_median:.0f} Jahre"),
    ("Alter Minimum", f"{alter_min:.0f} Jahre"),
    ("Alter Maximum", f"{alter_max:.0f} Jahre"),
    ("Alter Spannweite", f"{alter_range:.0f} Jahre"),
    ("Erfahrung mit Augmented Reality ‚Äì Ja", f"{tech_counts.get('Ja', tech_counts.get(1, 0)):.0f}%"),
    ("Erfahrung mit Augmented Reality ‚Äì Nein", f"{tech_counts.get('Nein', tech_counts.get(0, 0)):.0f}%"),
]

# Kaufverhalten hinzuf√ºgen
for kategorie in ["Sehr h√§ufig", "Oft", "Gelegentlich", "Selten"]:
    prozent = kauf_counts.get(kategorie, 0)
    rows.append((f"Kauft online: {kategorie}", f"{prozent:.0f}%"))

# DataFrame erstellen
auswertung = pd.DataFrame(rows, columns=["Merkmal", "Wert"])

# Ordner anlegen und Datei speichern
os.makedirs("Auswertung", exist_ok=True)
auswertung.to_excel("Auswertung/teilnehmerauswertung.xlsx", index=False)
```
